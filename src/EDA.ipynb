```python
df = pd.read_csv(config.combined_data_path)
df
```

```python
df['Newcol'] = (df['bnh_res_path'] + df['hsp1_res_path']) - (df['anh_res_path'] + df['wnh_res_path'])
```

```python
df.corr()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Newcol'], df['UHI Index'], alpha=0.5, s=2)

# 擬合一階線性回歸
coef = np.polyfit(df['Newcol'], df['UHI Index'], 1)
x = np.linspace(df['Newcol'].min(), df['Newcol'].max(), 100)
y = coef[0] * x + coef[1]
plt.plot(x, y, color='navy', linewidth=2)

plt.xlabel("(B+H) - (A+W)", fontsize=12)
plt.ylabel("UHI Index", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Add abbreviation key
annotation = (
    "B = Black\n"
    "H = Hispanic\n"
    "A = Asian\n"
    "W = White"
)
plt.gcf().text(
    0.14, 0.865,          # position: 2 % from left, 2 % from top
    annotation,
    va="top", ha="left",
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.3",
              facecolor="white", edgecolor="gray", alpha=0.7)
)
plt.savefig(os.path.join(config.figures_path, 'scatter_plot_2.png'), dpi=300, bbox_inches='tight')
plt.show()
```
